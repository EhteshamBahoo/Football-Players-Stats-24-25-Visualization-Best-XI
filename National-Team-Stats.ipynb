```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
file_path = 'football_stats/players_data-2024_2025.csv'
df = pd.read_csv(file_path)
```

```python
columns_needed = ['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'MP', 'Gls', 'Ast', 'Tkl', 'Int', 'Cmp%', 'Save%']
df = df[columns_needed].dropna(subset=['Nation']) 
```

```python
forwards = df[df['Pos'].str.contains('FW', na=False, case=False)]
midfielders = df[df['Pos'].str.contains('MF', na=False, case=False)]
defenders = df[df['Pos'].str.contains('DF', na=False, case=False)]

# **Step 4: Function to plot top 5 nations producing players**
def plot_top_nations(position_group, title, color):
    top_nations = position_group.groupby('Nation').size().sort_values(ascending=False).head(5)

    plt.figure(figsize=(8, 5))
    
    # Generate a gradient color from darkest to lightest
    palette = sns.color_palette(color, as_cmap=True)
    
    sns.barplot(
        x=top_nations.values, 
        y=top_nations.index, 
        hue=top_nations.index,  # Assigning hue for proper coloring
        palette=[palette(i / max(top_nations.values)) for i in top_nations.values],  
        legend=False
    )
    
    plt.xlabel("Number of Players")
    plt.ylabel("Country")
    plt.title(title, fontsize=14, fontweight='bold')
    plt.show()
```

```python
plot_top_nations(forwards, "Top 5 Countries Producing Forwards", "Reds")
plot_top_nations(midfielders, "Top 5 Countries Producing Midfielders", "Greens")
plot_top_nations(defenders, "Top 5 Countries Producing Defenders", "Blues")
```

## Top 6 Nations are:
1. France (FRA)
2. Spain (ESP)
3. Germany (GER)
4. England (ENG)
5. Italy (ITA)
6. Brazil (BRA)

```python
team_stats = df.groupby('Nation').agg({
    'Gls': 'sum',      # Total Goals
    'Ast': 'sum',      # Total Assists
    'Tkl': 'sum',      # Total Tackles
    'Cmp%': 'mean',    # Average Pass Completion
    'Save%': 'mean'    # Average Save %
}).dropna() 
```

```python
top_teams = team_stats.sort_values(by='Gls', ascending=False).head(6)
```

```python
plt.figure(figsize=(12, 6))
top_teams.plot(kind='bar', colormap='coolwarm', alpha=0.85, edgecolor='black')

plt.title('Top 6 National Teams - Performance Metrics', fontsize=14, fontweight='bold')
plt.xlabel('National Team', fontsize=12)
plt.ylabel('Performance Stats', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
top_teams['Rank'] = range(1, 7)
top_teams = top_teams.reset_index().rename(columns={
    'Nation': 'National Team',
    'Gls': 'Goals',
    'Ast': 'Assists',
    'Tkl': 'Tackles',
    'Cmp%': 'Pass Completion (%)',
    'Save%': 'Save Percentage (%)'
})
top_teams['Pass Completion (%)'] = top_teams['Pass Completion (%)'].apply(lambda x: f"{x:.2f}%")
top_teams['Save Percentage (%)'] = top_teams['Save Percentage (%)'].apply(lambda x: f"{x:.2f}%")

```

```python
plt.figure(figsize=(8, 4))
ax = plt.subplot(111, frame_on=False)

ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
table = plt.table(
    cellText=top_teams.values,
    colLabels=top_teams.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  # Scale for better readability

plt.title('Top 6 National Teams - Ranked', fontsize=14, fontweight='bold')
plt.show()
```

```python
file_path = 'football_stats/players_data-2024_2025.csv'
df = pd.read_csv(file_path)

columns_needed = ['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'MP', 'Gls', 'Ast', 'Tkl', 'Int', 'Cmp%', 'Save%']
df = df[columns_needed].dropna(subset=['Nation'])

forwards = df[df['Pos'].str.contains('FW', na=False, case=False)]
midfielders = df[df['Pos'].str.contains('MF', na=False, case=False)]
defenders = df[df['Pos'].str.contains('DF', na=False, case=False)]

top_teams = df.groupby('Nation').agg({
    'Gls': 'sum',
    'Ast': 'sum',
    'Tkl': 'sum',
    'Cmp%': 'mean',
    'Save%': 'mean'
}).dropna()

top_teams = top_teams.sort_values(by='Gls', ascending=False).head(6)
top_teams_nations = top_teams.index.tolist()

forwards_top_teams = forwards[forwards['Nation'].isin(top_teams_nations)]
midfielders_top_teams = midfielders[midfielders['Nation'].isin(top_teams_nations)]
defenders_top_teams = defenders[defenders['Nation'].isin(top_teams_nations)]

# Create a fixed color palette for the nations
fixed_palette = sns.color_palette("Set2", n_colors=len(top_teams_nations))
nation_to_color = dict(zip(top_teams_nations, fixed_palette))
```

```python
def plot_best_player(position_group, stat_col, title, nation_to_color):
    best_players = []

    for nation in top_teams_nations:
        nation_group = position_group[position_group['Nation'] == nation]
        best_player = nation_group.sort_values(by=stat_col, ascending=False).head(1)
        best_players.append(best_player)

    best_players_df = pd.concat(best_players)

    plt.figure(figsize=(10, 6))

    sns.set(style="whitegrid")

    bars = sns.barplot(
        x=best_players_df[stat_col],
        y=best_players_df['Player'],
        hue=best_players_df['Nation'],
        palette=[nation_to_color[nation] for nation in best_players_df['Nation']],
        dodge=False
    )

    for index, row in best_players_df.iterrows():
        flag_path = f"flags/{row['Nation']}.png"
        if os.path.exists(flag_path):
            img = mpimg.imread(flag_path)
            imagebox = OffsetImage(img, zoom=0.1)
            ab = AnnotationBbox(imagebox, (row[stat_col] + 0.5, index), frameon=False, boxcoords="data")
            bars.add_artist(ab)

    plt.xlabel(f"{stat_col} (Best Player from Each Nation)", fontsize=14)
    plt.ylabel("Player", fontsize=12)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

plot_best_player(forwards_top_teams, 'Gls', 'Best Forward from Each of Top 6 Countries', nation_to_color)
plot_best_player(midfielders_top_teams, 'Ast', 'Best Midfielder from Each of Top 6 Countries', nation_to_color)
plot_best_player(defenders_top_teams, 'Tkl', 'Best Defender from Each of Top 6 Countries', nation_to_color)
```
